```python
# exploratory data analysis
import pandas as pd
import seaborn as sns
```

```python
data=pd.read_csv('titanic.csv')
data
```

```python
# to show count plot of survived column
print(data['Survived'].value_counts())
sns.countplot(data['Survived'])
```

```python
sns.countplot(data['Embarked'])
```

```python
sns.countplot(data['Sex'])
```

```python
# using pie chart we can see the percentage 
data['Survived'].value_counts().plot(kind='pie',autopct='%.2f')
```

```python
# using histogram to see the numerical data visualize
import matplotlib.pyplot as plt
plt.hist(data['Age'])
```

```python
# help to show PDF to find probability of occurance of age 
sns.distplot(data['Age'])
```

```python
# box plot help to visualize the outlier and noisy data
sns.boxplot(data['Fare'])
```

```python
# plotting the numerical-numerical data column
sns.scatterplot(data['Age'],data['Fare'])
```

```python
# we can also distinct another column
sns.scatterplot(data['Age'],data['Fare'],hue=data['Sex'])
```

```python
# you can do multiple analysis using columns
sns.scatterplot(data['Age'],data['Fare'],hue=data['Sex'],style=data['Pclass'])
```

```python
# to show relation between pclass and fare using bar plot(num-cat)
sns.barplot(data['Pclass'],data['Fare'])
```

```python
sns.barplot(data['Pclass'],data['Fare'],hue=data['Sex'])
```

```python
sns.boxplot(data['Sex'],data['Age'])
```

```python
# it show the survived with respect to age
sns.distplot(data[data['Survived']==0]['Age'],hist=False)
sns.distplot(data[data['Survived']==1]['Age'],hist=False)
```

```python
sns.distplot(data[data['Survived']==0]['Fare'],hist=False)
sns.distplot(data[data['Survived']==1]['Fare'],hist=False)
```

```python
sns.heatmap(pd.crosstab(data['Survived'],data['Pclass']))
```

```python
pd.crosstab(data['Survived'],data['Pclass'])
```

```python
# to find percentage of sex that survived or not
(data.groupby('Sex').mean()['Survived']*100)
```

```python
data.groupby('Sex').mean()
```

```python
# To see visualization of the datset
sns.pairplot(data,hue='Survived')
```
